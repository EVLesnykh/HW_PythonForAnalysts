```python
# Подключаем библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1.1. Скачать следующие данные: kc-house-data

### 1.2. Изучите стоимости недвижимости

3. С ростом стоимости дома от 500 тыс. долларов количество представленных домов идет на спад.
4. Дома стоимостью выше 2 млн. долларов представлены в ограниченном количестве.

### Выводы:

### Подготовительный этап
Построим матрицу корреляций для того, чтобы найти линейные связи между признаками и соответственно выявить как один признак влияет на другой.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

### Выводы:
Согласно полученной матрицы корреляций, выявлены наиболее сильные связи между признаками, влияющие на стоимость дома.<br>

На стоимость дома существенно влияют следующие признаки (коэффициент корреляции от 0.5 до 1):
- количеством ванных комнат (bathrooms) - коэфф.корреляции = 0.5
- жилая площадь (sqft_living) - коэфф.корреляции = 0.7
- класс дома (grade) - коэфф.корреляции = 0.7
- площадь дома без учета площади подвала (sqft_above) - коэфф.корреляции = 0.6
- жилая площадь внутри жилья для ближайших 15 соседей (sqft_living15) - коэфф.корреляции = 0.6<br>

Оставшиеся признаки с коэфф.кореляции от 0 до 0.5 имеют слабое (несущественное) влияние на стоимость дома.

Итак, на основании полученных данных построим графики.<br>
P.S. Рассмотрим наиболее сильные связи между признаками.

### 1. Зависимость цены дома от класса дома

```python
sns.boxplot(x=df['grade'], y=df['price'], showfliers=False)
plt.title('Зависимость цены дома от класса дома')
plt.xticks(rotation=50);
```

### Выводы:
Цена дома сильно зависит от класса дома, чем выше класс тем выше цена и наоборот соответственно.

### 2. Зависимость цены дома от жилой площади дома

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

### Выводы:
Прослеживается прямая зависимость цены дома от жилой площади дома. Чем больше жилая площадь тем выше цена и наоборот.

### 3. Зависимость цены дома от количества ванных комнат

```python
sns.boxplot(x=df['bathrooms'], y=df['price'], showfliers=False)
plt.title('Зависимость цены дома от количества ванных комнат')
plt.xticks(rotation=50);
```

### Выводы:
Цена дома зависит от количества ванных комнат, чем больше ванных комнта тем выше цена и наоборот соответственно.

### 4. Зависимость цены дома от площади дома без учета площади подвала

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg');
```

### Выводы:
Прослеживается прямая зависимость цены дома от жилой площади дома без учета подвала. Чем больше площадь дома без учета подвала тем выше цена и наоборот.


### 5. Зависимость цены дома от жилой площади внутри жилья для ближайших 15 соседей

```python
sns.jointplot(x=df['price'], y=df['sqft_living15'], kind='reg');
```

### Выводы:
Прослеживается прямая зависимость цены дома от жилой площади внутри жилья для ближайших 15 соседей. Чем больше жилая площадь внутри жилья для ближайших 15 соседей тем выше цена и наоборот.
